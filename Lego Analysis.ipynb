```python
#The number of LEGO sets released year-on-year.
#This might tell us something about how LEGO's product offering has changed over time.
sets_by_year = sets.groupby("year").count()
sets_by_year
```

```python
sets_by_year["set_num"].head()
```

```python
sets_by_year["set_num"].tail()
```

```python
#Showing the number of LEGO releases on a line chart using Matplotlib.
plt.plot(sets_by_year.index, sets_by_year.set_num);
```

```python
#Data is from late 2023, to plot the full calendar years, you will have to exclude some data from chart.
plt.plot(sets_by_year.index[:-1], sets_by_year.set_num[:-1]);
```

From this, we can see that LEGO published less than 10 different sets per year during its first few years of operation. But by 2022 the company had grown spectacularly, releasing 1140 sets in year 2021 alone!

We also see that while the first 45 years or so, LEGO had some steady growth in its product offering, but it was really in the mid-1990s that the number of sets produced by the company increased dramatically! We also see a brief decline in the early 2000s and a strong recovery around 2005 in the chart. In the year 2022, there was a noticeable decline in certain sectors, which may have been influenced by the conclusion of COVID-19 lockdown measures and an increase in recreational activities outdoors.

```python
#Number of different themes shipped by year.
themes_by_year = sets.groupby("year").agg({"theme_id": pd.Series.nunique})
```

```python
themes_by_year.rename(columns = {"theme_id": "nr_themes"}, inplace=True)
themes_by_year.head()
```

```python
themes_by_year.tail()
```

```python
plt.plot(themes_by_year.index[:-1], themes_by_year.nr_themes[:-1]);
```

```python
plt.plot(themes_by_year.index[:-1], themes_by_year.nr_themes[:-1])
plt.plot(sets_by_year.index[:-1], sets_by_year.set_num[:-1]);
```

```python
#Number of themes and number of sets have very different scales. 
#Configuring plot to show two different axes on the same chart.

ax1 = plt.gca() #get first axis
ax2 = ax1.twinx() #create another axis that shares same x-axis

ax1.plot(themes_by_year.index[:-1], themes_by_year.nr_themes[:-1], color="g")
ax2.plot(sets_by_year.index[:-1], sets_by_year.set_num[:-1])

ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Themes", color="Green")
ax2.set_ylabel("Number of Sets", color="blue");
```

```python
#Fiding the average number of parts per set. How many parts did the average LEGO set released in 1949 compared to 2022?
parts_per_set = sets.groupby("year").agg({"num_parts": pd.Series.mean})
parts_per_set.head()
```

```python
parts_per_set.tail()
```

```python
#Showing how the size and complexity of LEGO sets increased over time based on the number of parts.
#Scatter plot show the average number of parts over time
plt.scatter(parts_per_set.index[:-1], parts_per_set.num_parts[:-1]);
```

From the chart, we can definitely make out an upward trend in the size and complexity of the LEGO sets based on the average number of parts. In the 2020s the average set contained around 200 individual pieces, which is roughly double what average LEGO set used to contain in the 1960s.

## Merging DataFrames based on key

```python
#Here I'm providing a dictionary to create the DataFrame. The keys in the dictionary become my column names.
sets_theme_count = pd.DataFrame({"id": sets_theme_count.index,
                              "set_count": sets_theme_count.values})
sets_theme_count.head()
```

```python
merged_df = pd.merge(sets_theme_count, themes, on="id")
merged_df[:5]
```

Star Wars is indeed the theme with the most LEGO sets. Let's plot the top 10 themes on a chart.

```python
plt.figure(figsize=(8,6))
plt.xticks(fontsize=12, rotation=60)
plt.yticks(fontsize=12)
plt.xlabel("Theme Name", fontsize=12)
plt.ylabel("Number of Sets", fontsize=12)


plt.bar(merged_df.name[:10], merged_df.set_count[:10]);
```
